2.Problem 02

hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.

Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.

```python
uploaded=files.upload()
```

```python
data1=pd.read_csv("LabTAT.csv")
```

```python
data1.head()
```

```python
data1.info()
```

```python
data1.describe()
```

```python
#ANOVA Testing

#null hypothesis : there is no difference in average turn around time of reports of the laboratories on their preferred list

#alternate hypothesis : there is  difference in average turn around time of reports of the laboratories on their preferred list

#as there are multiple continues variable i am choosing Anova test for predicting the null and alternate hypothesis

#Fscore and probability are as fallows:F_onewayResult(statistic=118.70421654401437, pvalue=2.1156708949992414e-57)

'''Anova Testing  scores are as fallows'''

'''F-score : 118.70421654401437
    probability values : 2.1156708949992414e-57 '''

from scipy import stats
F_score,p_values=stats.f_oneway(data1.iloc[:,0], data1.iloc[:,1],data1.iloc[:,2],data1.iloc[:,3])

print('F-score :',F_score)
print('probability values :',p_values)
'''
Conclusion :
            Accepting the Alternative hypothesis
            comparing the probability with level of significance where probability is around 0% '''
```

3.Problem 03

Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions

```python
uploaded=files.upload()
```

```python
data2=pd.read_csv("BuyerRatio.csv")
```

```python
data2.head()
```

```python
data2.describe()
```

```python
#CHI-square Test

#Null hypothesis : male- female buyer ratio are similar across region

#Alternate hypothesis : male-female buyer ratio are not similar across region

#calculating the chisquare  test

chi_square,p_values,dof,expected = stats.chi2_contingency(data2.iloc[:,1:])

print('chi_square test values :',chi_square)
print('pvalues:',p_values)
```

```python
#test values and probability of the above problem is as follows
'''
    chi_square test values : 1.595945538661058

    pvalues: 0.6603094907091882'''

'''conclusion : unable to reject null hypothesis beacuse pvalues is greater then level of significance(ie.p>alpha,p=66%,alpha=5%)it states all proportion of sales are equal'''
```

TeleCall uses 4 centers around the globe to process customer order forms. They audit a certain % of the customer order forms. Any error in order form renders it defective and has to be reworked before processing. The manager wants to check whether the defective % varies by centre. Please analyze the data at 5% significance level and help the manager draw appropriate inferences

```python
uploaded=files.upload()
```

```python
data3=pd.read_csv("Costomer+OrderForm.csv")
```

```python
data3.head()
```

```python
data3.describe()
```

```python
data3.info()
```

```python
#here all the data present in the datasets are in the form of categorical

#from the data sets i found two different values of classification (ie.Error free  and defective ) in all those variables

#lets create the new datasets by considering two classification

new_df=pd.DataFrame({'Phillippines':data3.Phillippines.value_counts(),'Indonesia':data3.Indonesia.value_counts(),
                     'Malta':data3.Malta.value_counts(),'India':data3.India.value_counts()})

new_df
```

```python
#CHI-Square test

#Null hypothesis : defective percentage around the globe to process the customer order form will not varies

#Alternate hypothesis : defective percentage around the globe to process the customer order form will varies

#calculating the chisquare  test

chi_square,p_values,dof,expected = stats.chi2_contingency(new_df)

print('chi_square test values :',chi_square)
print('pvalues:',p_values)

```

```python
#test values and probability of the above problem is as fallows

'''   chi_square test values : 3.858960685820355

      pvalues: 0.2771020991233135 '''

'''conclusion : unable to reject null hypothesis beacuse pvalues is greater then level of significance(ie.p>alpha,p=27.7%,alpha=5%)

                defective percentage not varies '''

```
